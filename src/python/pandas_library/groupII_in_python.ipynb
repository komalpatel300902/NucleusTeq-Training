```python
import pandas as pd
import numpy as np
```

```python
name,field1,field2,field3 = [],[],[],[]

for w in range(1,10):
    for x in range(20,50,10):
        for y in range(-100,-400,-100):
            for z in range(1000,4000,1000):
                name.append(str(w))
                field1.append(x)
                field2.append(y)
                field3.append(z)
```

```python
df = pd.DataFrame({
    "name":name,
    "field1":field1,
    "field2":field2,
    "field3":field3
})
```

```python
data1 = {'Name': ['Alice', 'Bob', 'Charlie', 'David'] * 5,
        'Age': [25, 30, 35, 40] * 5,
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'] * 5}

df1 = pd.DataFrame(data1)
```

```python
data2 = {'Name': ['Alice', 'Bob', 'Charlie', 'David'] * 5,
        'Id': [25, 30, 35, 40] * 5,
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'] * 5}

df2 = pd.DataFrame(data2)
```

```python
merge_df = pd.merge(df1,df2,how = "inner",left_on = "Age",right_on = "Id")
```

```python
merge_df
```

```python
merge_df.drop_duplicates(subset = ["Name_x"])
```
