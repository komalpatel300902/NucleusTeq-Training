## 5
Create a new column in each table called "Age_Group" and categorize the gods/goddesses into groups such as "Young" (age < 5000), "Middle-aged" (age between 5000 and 8000), and "Old" (age > 8000).


```python
df1
```

```python
def age_group(rows):
    if rows["Age"] < 5000:
        rows["Age_Group"] = "Young"
    elif rows["Age"] >= 5000 and rows["Age"] <= 8000:
        rows["Age_Group"] = "Middle-aged"
    elif rows["Age"] > 8000:
        rows["Age_Group"] = "Old"
    return rows

df1.apply(age_group,axis = 1)
```

```python
df2.apply(age_group,axis = 1)
```

## 7
Write a Python program using for loop to iterate over the "Age" column of the merged table (after merging the gods and goddesses tables) and print out the names of gods/goddesses who are older than 8000 years.

```python
gods = main_df["God"].replace(to_replace = np.nan,value = "")
goddess = main_df["Goddess"].replace(to_replace = np.nan,value = "")
age = main_df["Age"].tolist()
length = len(age)
for x in range(length):
    if age[x] > 8000:
        print(gods[x],end = "")
        print(goddess[x],end = ", ")
        print(age[x])
        
```

## 8
Write a Python program to find the oldest god/goddess from the merged table (after merging the gods and goddesses tables) by iterating through the "Age" column using a while loop. Print out the name of the oldest god/goddess and their age.

```python
gods = main_df["God"].replace(to_replace = np.nan,value = "")
goddess = main_df["Goddess"].replace(to_replace = np.nan,value = "")
age = main_df["Age"].tolist()

name = ""
ag = 0
index = 0
condition = True
length = len(age) -1
while condition:
    if index == length:
        condition = False
    if age[index] > ag:
        name = gods[index] + goddess[index]
        ag = age[index]
    index += 1
print(name,ag)
```
