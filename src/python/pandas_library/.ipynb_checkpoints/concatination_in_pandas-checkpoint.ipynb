```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    "name":["komal","nitin","Rahul","Reetik","Ramesh", "Raju","Suresh","mukesh","rakhi","jatin","rishab"],
    "age":[23,34,23,43,64,34,26,35,26,19,12]
})
```

```python
df2 = pd.DataFrame({
    "name":["MS Dhoni", "Jateja","Virat","Shubham"],
    "age":[42,41,39,30]
})
```

### concatination on axis 0

```python
df3= pd.concat([df1,df2],axis = 0)
```
